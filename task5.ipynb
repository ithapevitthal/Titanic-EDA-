```python
numeric_cols=["Age","Fare","SibSp","Parch","Survived"]
```

```python
data=df[numeric_cols].dropna(subset=["Age"])
```

```python
sns.pairplot(df,hue="Survived",diag_kind="hist")
plt.show()
```

```python
plt.figure(figsize=(6,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.histplot(data=df, x="Age", bins=30, hue="Survived", kde=True, multiple="stack")
plt.title("Age Distribution by Survival")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.histplot(data=df, x="Fare", bins=30, hue="Survived", kde=True, multiple="stack")
plt.title("Fare Distribution by Survival")
plt.show()
```

#BOXPLOT

```python

plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Survived", y="Age")
plt.title("Age vs Survival (Boxplot)")
plt.show()

```

```python

plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Survived", y="Fare")
plt.title("Fare vs Survival (Boxplot)")
plt.show()
```

```python
data1=data.head(10)
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived", style="Survived")
plt.title("Age vs Fare by Survival")
plt.show()


```

```python

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="SibSp", y="Parch", hue="Survived", size="Fare")
plt.title("Family Size vs Survival")
plt.show()
```

```python
print("OBSERVATIONS:")
print("1. Higher fares linked to better survival chances.")
print("2. Younger passengers (children) survived more often.")
print("3. Small families had better survival; large families worse.")
print("4. Higher Pclass (3rd class) strongly linked to lower survival.")
print("5. Age had weak correlation, but children stood out as survivors.")

print("\nSUMMARY:")
print(" Socio-economic status (Fare & Pclass) had a strong effect on survival.")
print(" Family size influenced chances: small families = better.")
print(" Women & children had better chances (from Sex column).")
print(" Overall: Higher fare, 1st/2nd class, and younger passengers survived more.")
```
